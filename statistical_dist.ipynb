```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define Function
def ecdf(df):
    """ Return x, y array for plot CDF """
    x = np.sort(df)
    y = np.linspace(0, 1, len(df))
    return x, y
```

# Probability Distritution Funcions
## Discrete
Probability mass function (PMF)

## Continuous
Probability density function (PDF)

Cumulative distribution function (CDF)

```python
# Insert Seed and create a normal distritution sample array
np.random.seed(49)
normal_sample = np.random.normal(0,1, size = 10000)
sample_mean = np.mean(normal_sample)
sample_std = np.std(normal_sample)
print('Mean of sample is {}, and std dev is {}'.format(sample_mean, sample_std))
```

```python
plt.hist(normal_sample, bins = 100)
plt.title('Probability Mass Function')
plt.ylabel('PMF')
plt.xlabel('n')
plt.show();
```

```python
hist, bins = np.histogram(normal_sample, bins= 100, normed = True)
bin_centers = (bins[1:] + bins[:-1])*0.5
plt.plot(bin_centers, hist)
plt.title('Probability Density Function')
plt.ylabel('PDF')
plt.xlabel('n')
plt.show();
```

Poisson Distribution
- Discrete distribution
- Describes the number of events occuring in a fixed time interval or region of opportunity
- Requires only one parameter, lamda
- Bound by 0 and infinity

```python
wine_sample = np.random.poisson(12, 100000)
x = np.sort(wine_sample)
y = np.linspace(0,1,len(wine_sample))
```

```python
plt.plot(x, y, marker = '.', linestyle = 'none')
plt.ylabel('CDF')
plt.xlabel('sale in a day')
plt.show()
```
