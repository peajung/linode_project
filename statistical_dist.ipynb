```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample = np.random.binomial(10, 0.08, size = 10000000)
```

```python
x = np.sort(sample)
y = np.linspace(0,1, 10000000)
```

```python
plt.plot(x, y, marker = '.', linestyle = 'none')
plt.ylabel('CDF')
plt.xlabel('no color blind')
plt.show()
```

Poisson Distribution
- Discrete distribution
- Describes the number of events occuring in a fixed time interval or region of opportunity
- Requires only one parameter, lamda
- Bound by 0 and infinity

```python
wine_sample = np.random.poisson(12, 100000)
x = np.sort(wine_sample)
y = np.linspace(0,1,len(wine_sample))
```

```python
plt.plot(x, y, marker = '.', linestyle = 'none')
plt.ylabel('CDF')
plt.xlabel('sale in a day')
plt.show()
```

```python
plt.hist(wine_sample, bins= int(np.sqrt(len(wine_sample))))
plt.show()
```
